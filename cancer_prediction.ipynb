```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_excel('cancer patient data sets.xlsx')
df.head()
```

```python
df.columns
```

```python
x = df[['Age', 'Gender' , 'Alcohol use',
       'chronic Lung Disease', 'Obesity', 'Smoking',
        'Chest Pain', 'Coughing of Blood', 'Fatigue',
       'Weight Loss', 'Shortness of Breath',
       'Swallowing Difficulty',  'Frequent Cold',
       'Dry Cough', 'Snoring']]

y = df['Level']
```

```python
x.head()
```

```python
y.head()
```

```python
y.value_counts()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
# Create an instance of LabelEncoder and fit
encoder = LabelEncoder()
encoder.fit(y)
encoder.classes_
```

```python
# Transform y
encoded_y = encoder.transform(y)
print(encoded_y)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, encoded_y, test_size=0.2)
```

```python
print ('Train set:', x_train.shape,  y_train.shape)
print ('Test set:', x_test.shape,  y_test.shape)
```

```python
from sklearn import svm
from sklearn.metrics import classification_report, f1_score, jaccard_score
```

```python
for kernel in ['linear','poly','sigmoid','rbf']:
    model = svm.SVC(kernel=kernel).fit(x_train, y_train)
    pred=model.predict(x_test)
    print('The classification report of svm of kernel =',kernel,'is :\n')
    print(classification_report(pred,y_test),'\n')

    print("Avg F1-score: %.4f" % f1_score(y_test, pred, average='weighted',zero_division="warn"))
    print("Jaccard score: %.4f" % jaccard_score(y_test, pred, average='micro'),'\n\n')
```

```python
from joblib import dump, load
```

```python
svm_model = svm.SVC(kernel='linear').fit(x_train, y_train)
dump(svm_model, 'svm_model.joblib')
```

```python
load_model = load('svm_model.joblib')
pred=load_model.predict(x_test)
print(classification_report(pred,y_test),'\n')
```

```python
import numpy as np

def predict_health_risk():
    a = int(input("Enter Your Age: "))
    b = int(input("Enter Your Gender (1-male, 2-female): "))
    print("Please Enter The reading from 1-9\n")
    c = int(input("Rate Your Alcohol Consumption: "))
    d = int(input("Chronic Lung Disease value: "))
    e = int(input("Rate Your Obesity: "))
    f = int(input("Rate Your Smoking Addiction: "))
    g = int(input("Chest Pain Level: "))
    h = int(input("Rate Your Coughing of Blood: "))
    i = int(input("Rate Your Fatigue: "))
    j = int(input("Rate Your Weight Loss: "))
    k = int(input("Shortness of Breath: "))
    l = int(input("Swallowing Difficulty: "))
    m = int(input("Frequent Cold: "))
    n = int(input("Dry Cough: "))
    o = int(input("Snoring: "))

    input_data = (a, b, c, d, e, f, g, h, i, j, k, l, m, n, o)
    input_array = np.asarray(input_data)
    input_reshape = input_array.reshape(1, -1)
    pred = load_model.predict(input_reshape)
    encoded_y = encoder.inverse_transform(pred)
    print("Health Risk Level:", encoded_y)
    print("Predicted Values:", pred)

predict_health_risk()
while True:
    choice = input("Enter 1 to execute again, 2 to exit: ")
    if choice == '1':
        predict_health_risk()
    elif choice == '2':
        print("Thank you for using this program. Goodbye!")
        break
    else:
        print("Invalid choice. Please enter 1 to execute again or 2 to exit.")
```
